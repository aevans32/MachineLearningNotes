#**14. data.groupby()**
Este cdigo agrupa el DataFrame por los valores nicos de una columna. Pero por s solo no hace nada visible hasta que le agregas una operacin, como .sum(), .mean(), .count(), etc.

```python
# Aplicar el metodo a la columna renewal, solo nos genera un codigo de memoria.
data.groupby('renewal')
```

```python
# Aqui estamos agrupando por 2 columnas, 'residence_area_type' y 'renewal'. Esto muestra el promedio de primas segun tipo de zona y si es que el cliente renovo o no.
data.groupby(['residence_area_type', 'renewal'])['premium'].mean().apply(lambda x: f'{x:,.2f}')
```

```python
# Asi mismo podemos usar describe() para realizar un analisis estadistico completo por cada columna de las 2 agrupaciones resultantes.
data.groupby('renewal').describe()
```

#**15. One Hot Encoding**
One Hot Encoding es una tcnica utilizada para convertir variables categricas (de tipo texto o object) en variables numricas que pueden ser utilizadas por algoritmos de machine learning.

Consiste en crear una nueva columna para cada categora distinta de la variable original, donde cada fila indica con 1 o 0 si pertenece a esa categora.

**Qu hace esta lnea?**

 * columns=['sourcing_channel', 'residence_area_type']

Especifica las columnas categricas que se desean codificar.

---

 * drop_first=True

Elimina la primera categora de cada variable para evitar colinealidad (es decir, que una categora sea perfectamente predecible a partir de las dems). Esto es til especialmente en modelos lineales.

---

 * dtype=int

Asegura que las columnas generadas sean de tipo entero (0 o 1), en lugar de booleano (True/False).

```python
data_encoded = pd.get_dummies(data, columns=['sourcing_channel', 'residence_area_type'], drop_first=True, dtype=int)
```

```python
data_encoded.head()
```

#**16. Escalar variables**
El escalado es una tcnica que transforma las variables numricas para que estn en una escala comparable.

Esto es especialmente importante en modelos sensibles a la magnitud de los datos, como regresin logstica, redes neuronales, SVM y PCA.

Uno de los mtodos ms comunes es el escalado estndar (Standard Scaling), que transforma cada columna numrica para que tenga:

 * Media = 0
 * Desviacin estndar = 1

**Qu hace cada parte?**

 * StandardScaler()

Crea un objeto que realizar la transformacin estndar de los datos.

---

 * select_dtypes(include='number')

Filtra solo las columnas numricas (int64, float64), evitando errores con columnas no numricas o dummies booleanas.

---

 * scaler.fit_transform(...)

Ajusta el escalador a los datos y luego los transforma usando:

z = (  ) / 

Donde:
*  es el valor original,
*  es la media de la columna,
*  es la desviacin estndar.

---

 * data_scaled

Es una nueva copia del DataFrame original data_encoded, pero con todas las columnas numricas escaladas. Las columnas codificadas (One Hot) tambin se transforman a escala 0-1, aunque en modelos como rboles no siempre es necesario.

---

***Para qu sirve?***

Mejora la convergencia de los algoritmos de aprendizaje automtico.

Asegura que ninguna variable domine por tener un rango mayor.

Facilita anlisis de correlacin, PCA y clustering.




```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Seleccionamos solo las columnas numericas
numeric_columns = data_encoded.select_dtypes(include='number').columns

# Copiamos la data a una nueva variable
data_scaled = data_encoded.copy()
# y aplicamos el escalado
data_scaled[numeric_columns] = scaler.fit_transform(data_encoded[numeric_columns])
```

```python
data_scaled.head()
```

#**17. Visualizacion de correlacion usando sns.heatmap()**
Una matriz de correlacin muestra cmo se relacionan numricamente todas las variables entre s. Es especialmente til para:

* Identificar relaciones lineales fuertes o dbiles,
* Detectar multicolinealidad,
* Guiar la seleccin de variables en modelos predictivos.

---

**Qu hace cada lnea?**

* plt.figure(figsize=(12,10))

Define el tamao de la figura del grfico (12 de ancho  10 de alto).

---

* data_scaled.corr()

Calcula la matriz de correlacin de todas las columnas numricas del DataFrame escalado.
La correlacin vara entre:

+1  correlacin positiva perfecta,

0  sin correlacin lineal,

1  correlacin negativa perfecta.

---

* sns.heatmap(..., annot=True, cmap='coolwarm', fmt='.2f')

Genera el heatmap con:

annot=True  muestra los valores numricos en cada celda.

cmap='coolwarm'  usa un degradado de colores que va del rojo (negativo) al azul (positivo).

fmt='.2f'  muestra los nmeros con 2 decimales.

---

* plt.title(...) y plt.show()

Aade un ttulo descriptivo y muestra el grfico.


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sns.heatmap(data_scaled.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlacion entre variables escaladas')
plt.show()
```
