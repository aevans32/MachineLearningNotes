#**14. data.groupby()**
Este código agrupa el DataFrame por los valores únicos de una columna. Pero por sí solo no hace nada visible hasta que le agregas una operación, como .sum(), .mean(), .count(), etc.

```python
# Aplicar el metodo a la columna renewal, solo nos genera un codigo de memoria.
data.groupby('renewal')
```

```python
# Aqui estamos agrupando por 2 columnas, 'residence_area_type' y 'renewal'. Esto muestra el promedio de primas segun tipo de zona y si es que el cliente renovo o no.
data.groupby(['residence_area_type', 'renewal'])['premium'].mean().apply(lambda x: f'{x:,.2f}')
```

#**15. One Hot Encoding**
One Hot Encoding es una técnica utilizada para convertir variables categóricas (de tipo texto o object) en variables numéricas que pueden ser utilizadas por algoritmos de machine learning.

Consiste en crear una nueva columna para cada categoría distinta de la variable original, donde cada fila indica con 1 o 0 si pertenece a esa categoría.

**¿Qué hace esta línea?**

 * columns=['sourcing_channel', 'residence_area_type']

Especifica las columnas categóricas que se desean codificar.

---

 * drop_first=True

Elimina la primera categoría de cada variable para evitar colinealidad (es decir, que una categoría sea perfectamente predecible a partir de las demás). Esto es útil especialmente en modelos lineales.

---

 * dtype=int

Asegura que las columnas generadas sean de tipo entero (0 o 1), en lugar de booleano (True/False).

```python
data_encoded = pd.get_dummies(data, columns=['sourcing_channel', 'residence_area_type'], drop_first=True, dtype=int)
```

```python
data_encoded.head()
```

#**16. Escalar variables**
El escalado es una técnica que transforma las variables numéricas para que estén en una escala comparable.

Esto es especialmente importante en modelos sensibles a la magnitud de los datos, como regresión logística, redes neuronales, SVM y PCA.

Uno de los métodos más comunes es el escalado estándar (Standard Scaling), que transforma cada columna numérica para que tenga:

 * Media = 0
 * Desviación estándar = 1

**¿Qué hace cada parte?**

 * StandardScaler()

Crea un objeto que realizará la transformación estándar de los datos.

---

 * select_dtypes(include='number')

Filtra solo las columnas numéricas (int64, float64), evitando errores con columnas no numéricas o dummies booleanas.

---

 * scaler.fit_transform(...)

Ajusta el escalador a los datos y luego los transforma usando:

z = ( 𝑥−𝜇 ) / 𝜎

Donde:
* 𝑥 es el valor original,
* 𝜇 es la media de la columna,
* 𝜎 es la desviación estándar.

---

 * data_scaled

Es una nueva copia del DataFrame original data_encoded, pero con todas las columnas numéricas escaladas. Las columnas codificadas (One Hot) también se transforman a escala 0-1, aunque en modelos como árboles no siempre es necesario.

---

***¿Para qué sirve?***

Mejora la convergencia de los algoritmos de aprendizaje automático.

Asegura que ninguna variable domine por tener un rango mayor.

Facilita análisis de correlación, PCA y clustering.




```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Seleccionamos solo las columnas numericas
numeric_columns = data_encoded.select_dtypes(include='number').columns

# Copiamos la data a una nueva variable
data_scaled = data_encoded.copy()
# y aplicamos el escalado
data_scaled[numeric_columns] = scaler.fit_transform(data_encoded[numeric_columns])
```

```python
data_scaled.head()
```

#**17. Visualizacion de correlacion usando sns.heatmap()**
Una matriz de correlación muestra cómo se relacionan numéricamente todas las variables entre sí. Es especialmente útil para:

* Identificar relaciones lineales fuertes o débiles,
* Detectar multicolinealidad,
* Guiar la selección de variables en modelos predictivos.

---

**¿Qué hace cada línea?**

* plt.figure(figsize=(12,10))

Define el tamaño de la figura del gráfico (12 de ancho × 10 de alto).

---

* data_scaled.corr()

Calcula la matriz de correlación de todas las columnas numéricas del DataFrame escalado.
La correlación varía entre:

+1 → correlación positiva perfecta,

0 → sin correlación lineal,

–1 → correlación negativa perfecta.

---

* sns.heatmap(..., annot=True, cmap='coolwarm', fmt='.2f')

Genera el heatmap con:

annot=True → muestra los valores numéricos en cada celda.

cmap='coolwarm' → usa un degradado de colores que va del rojo (negativo) al azul (positivo).

fmt='.2f' → muestra los números con 2 decimales.

---

* plt.title(...) y plt.show()

Añade un título descriptivo y muestra el gráfico.


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sns.heatmap(data_scaled.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlacion entre variables escaladas')
plt.show()
```
